```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.array([0.195484, 0.199424, 0.140287, 0.117481, 0.0945779, 0.0883637])
p = np.arange(1, len(T) + 1)
T0 = T[0]
S = T0 / T
E = S / p
```

```python
plt.figure(figsize=[18, 10])
plt.title("Зависимость ускорения от числа процессов")
plt.plot(p, S)
plt.xlabel("p")
plt.ylabel("S")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=[18, 10])
plt.title("Зависимость эффективности от числа процессов")
plt.plot(p, E)
plt.xlabel("p")
plt.ylabel("E")
plt.grid()
plt.show()
```
